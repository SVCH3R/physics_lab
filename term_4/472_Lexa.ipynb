```python

plt=reload(plt)
f_ot_x = r**2 #
yerr = 2 * (0.2 / r) * r**2
x = np.array(range(6))
ox = x
oy = f_ot_x
w = 1.0 / (yerr**2)

plt.errorbar(
    ox,  # ординаты точек
    oy,  # абсциссы точек
    linestyle='',
    xerr = 0,
    yerr = yerr,
    marker=".",
    #labrl="T(delta)"
    elinewidth=1,
    color = "g",
)

coeff = np.polyfit(ox[:], oy[:], 1, cov=True, w=w)
p1 = np.poly1d(coeff[0])
plt.plot(ox[:], p1(ox[:]), linewidth=0.7, label='$r^{2}(n)$', color='firebrick', linestyle = '--')


plt.xlabel('n')
plt.ylabel('$r^{2}, cm^2$')


plt.suptitle('$r^2(n)$', fontsize=18)



plt.legend(loc='upper left')
plt.plot(figsize=(3, 3))
plt.ticklabel_format(scilimits=(5, 2))
plt.grid(True)
plt.show()

print(coeff)

k = coeff[0][0]
sigmak = coeff[1][0][0]

print(k, sigmak)
```
