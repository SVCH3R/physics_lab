```python
v = U_main
b = U_pauwlgkhi
y = v
x = b**2
# plt.ylabel()
# plt.xlabel()
# plt.title()
plt.grid(True)
dx = 0.1
dy = 0.004
plt.errorbar(x, y, xerr=dx, yerr=dy, fmt='o', color='blue', capsize=3)
pol = np.polyfit(x, y, deg=1, cov=True)
print(pol)
pol = pol[0]
yfit = np.polyval(pol, x)
plt.plot(x, yfit, color="firebrick")
plt.legend(loc='best', fontsize=12)
plt.show()
```
